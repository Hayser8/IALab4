### 1. ¿Qué son los grafos computacionales y cuál es su importancia en el cálculo de gradientes en backpropagation?

Un **grafo computacional** es una representación estructurada de una función matemática donde los nodos representan operaciones o variables y las aristas representan el flujo de datos entre estas operaciones. En el contexto del cálculo de gradientes y el algoritmo de **backpropagation**, los grafos computacionales permiten descomponer la función de pérdida en una secuencia de operaciones elementales, facilitando el cálculo eficiente de derivadas mediante la **diferenciación automática**.

La importancia de los grafos computacionales en **backpropagation** radica en que permiten:

- **Optimización del cálculo de gradientes**: Evitan la redundancia en el cálculo de derivadas utilizando el método de la propagación inversa.
- **Eficiencia computacional**: Permiten almacenar los resultados intermedios de cada operación para ser reutilizados en la propagación inversa.
- **Facilidad en la implementación de modelos de aprendizaje profundo**: Frameworks como TensorFlow y PyTorch utilizan grafos computacionales para representar redes neuronales, facilitando la computación en hardware optimizado (GPU).

### 2. Componentes y pasos de una red neuronal. ¿Cómo mejorar el perceptrón del laboratorio pasado?

#### Componentes de una red neuronal:
Una red neuronal consta de los siguientes elementos fundamentales:

1. **Neuronas**: Unidades básicas que reciben una entrada, la procesan y generan una salida.
2. **Pesos y Sesgos**: Parámetros ajustables que permiten que la red aprenda patrones.
3. **Funciones de activación**: Aplicadas a la salida de las neuronas para introducir no linealidad (Ejemplo: ReLU, Sigmoide, Tanh).
4. **Capas**:
   - **Capa de entrada**: Recibe los datos.
   - **Capas ocultas**: Procesan la información mediante combinaciones lineales y funciones de activación.
   - **Capa de salida**: Genera la predicción final.
5. **Función de pérdida**: Mide la diferencia entre la salida real y la predicción.
6. **Optimizador**: Ajusta los pesos mediante **descenso de gradiente** para minimizar la pérdida.

#### Pasos en el entrenamiento de una red neuronal:
1. Inicializar los pesos y sesgos.
2. Realizar la **propagación hacia adelante** (Forward Pass) para obtener la salida.
3. Calcular la **función de pérdida**.
4. Aplicar **backpropagation** para calcular los gradientes.
5. **Actualizar los pesos** usando un optimizador como **Adam** o **SGD**.
6. Repetir los pasos hasta la convergencia.

### 3. Selección del valor K en K-Means usando el método de la silueta

El **método de la silueta** es una técnica utilizada para determinar el número óptimo de clusters (**K**) en el algoritmo de **K-Means** midiendo qué tan bien están agrupados los puntos dentro de sus clusters en comparación con otros clusters.

#### **Fórmula de la puntuación de silueta**:
Para cada punto \( i \) en el conjunto de datos:

1. **Calcular \( a(i) \)**: Distancia promedio entre \( i \) y todos los demás puntos en su mismo cluster.
2. **Calcular \( b(i) \)**: Distancia promedio entre \( i \) y los puntos en el cluster más cercano (al que no pertenece).
3. **Calcular la silueta de \( i \)**:
$$
s(i) = \frac{b(i) - a(i)}{\max(a(i), b(i))}
$$

   Donde:
   - \( s(i) \) varía entre -1 y 1.
   - Un valor cercano a 1 indica que el punto está bien agrupado.
   - Un valor cercano a 0 indica que el punto está en la frontera entre clusters.
   - Un valor negativo indica que el punto está mal agrupado.

#### **Elección de K**:
Se calcula la **puntuación media de silueta** para diferentes valores de \( K \), y se selecciona el \( K \) que maximiza esta puntuación.

**Asunciones**:
- Se asume que los datos pueden dividirse en clusters bien definidos.
- Se asume que la distancia euclidiana es una buena métrica para medir la cohesión de los clusters.

### 4. PCA y K-Means: ¿Cómo puede ayudar PCA a mejorar la calidad de los clusters?

**Principal Component Analysis (PCA)** es un método de reducción de dimensionalidad que transforma los datos a un nuevo espacio con menos dimensiones manteniendo la mayor variabilidad posible.

#### **Cómo ayuda PCA en K-Means**:
1. **Reducción de ruido y redundancia**: PCA elimina dimensiones poco informativas, lo que ayuda a K-Means a encontrar clusters más claros.
2. **Manejo de la "maldición de la dimensionalidad"**: K-Means puede tener problemas en espacios de alta dimensión. PCA reduce la dimensionalidad, facilitando la agrupación.
3. **Mejor visualización**: Permite proyectar los datos en 2D o 3D para visualizar mejor los clusters formados.
4. **Aceleración del algoritmo**: Con menos dimensiones, K-Means converge más rápido y consume menos recursos computacionales.

### Implementación de K-Means desde Cero

```python
import numpy as np

class KMeansCustom:
    def __init__(self, n_clusters, max_iters=100, tol=1e-4):
        self.n_clusters = n_clusters
        self.max_iters = max_iters
        self.tol = tol
        self.centroids = None

    def fit(self, X):
        np.random.seed(42)
        random_indices = np.random.choice(len(X), self.n_clusters, replace=False)
        self.centroids = X[random_indices]
        
        for _ in range(self.max_iters):
            clusters = self._assign_clusters(X)
            new_centroids = np.array([X[clusters == k].mean(axis=0) for k in range(self.n_clusters)])
            
            if np.linalg.norm(self.centroids - new_centroids) < self.tol:
                break
            
            self.centroids = new_centroids

    def _assign_clusters(self, X):
        distances = np.linalg.norm(X[:, np.newaxis] - self.centroids, axis=2)
        return np.argmin(distances, axis=1)
    
    def predict(self, X):
        return self._assign_clusters(X)

X = df_scaled.values

kmeans_custom = KMeansCustom(n_clusters=5)
kmeans_custom.fit(X)

labels = kmeans_custom.predict(X)

df_scaled['Cluster'] = labels

print(df_scaled.head())
```

# **Selección del Número Óptimo de Clusters para Mixture Models**

## **1️⃣ Análisis de la Gráfica de BIC y AIC**
El gráfico muestra la evolución de los valores de **BIC (Bayesian Information Criterion)** y **AIC (Akaike Information Criterion)** a medida que aumentamos el número de clusters.

📌 **Puntos clave observados:**
- Tanto **BIC como AIC disminuyen rápidamente** en los primeros valores de clusters, indicando que más clusters mejoran el ajuste del modelo.
- La **reducción más fuerte ocurre entre 2 y 6 clusters**.
- A partir de **6 clusters**, la disminución de BIC y AIC se vuelve más lenta, lo que sugiere que agregar más clusters no mejora significativamente el modelo.

## **2️⃣ Selección del Número Óptimo de Clusters**
Para elegir el número óptimo de clusters, seguimos los siguientes criterios:
- **El punto donde BIC y AIC dejan de disminuir rápidamente** indica el mejor equilibrio entre ajuste y simplicidad.
- Observamos que **K = 6** parece ser el punto donde la mejora comienza a estabilizarse.
- Elegir un número mayor de clusters aumentaría la complejidad sin una mejora significativa.

✅ **Selección final:** **Usaremos 6 clusters para Mixture Models.**



```python
import numpy as np

class GaussianMixtureCustom:
    def __init__(self, n_components, max_iters=100, tol=1e-4, reg_covar=1e-6):
        self.n_components = n_components
        self.max_iters = max_iters
        self.tol = tol
        self.reg_covar = reg_covar
        self.means = None
        self.covariances = None
        self.weights = None

    def initialize_parameters(self, X):
        np.random.seed(42)
        n_samples, n_features = X.shape
        
        self.means = X[np.random.choice(n_samples, self.n_components, replace=False)]
        self.covariances = np.array([np.cov(X.T) + np.eye(n_features) * self.reg_covar for _ in range(self.n_components)])
        self.weights = np.full(self.n_components, 1 / self.n_components)

    def gaussian_pdf(self, X, mean, cov):
        n = X.shape[1]
        cov_reg = cov + np.eye(n) * self.reg_covar  
        cov_det = np.linalg.det(cov_reg)
        cov_inv = np.linalg.inv(cov_reg)
        norm_const = 1.0 / (np.sqrt((2 * np.pi) ** n * cov_det))
        diff = X - mean
        exponent = np.einsum('...j,jk,...k->...', diff, cov_inv, diff)
        return norm_const * np.exp(-0.5 * exponent)

    def expectation_step(self, X):
        responsibilities = np.zeros((X.shape[0], self.n_components))
        for k in range(self.n_components):
            responsibilities[:, k] = self.weights[k] * self.gaussian_pdf(X, self.means[k], self.covariances[k])
        
        sum_responsibilities = responsibilities.sum(axis=1, keepdims=True) + self.reg_covar 
        responsibilities /= sum_responsibilities
        return responsibilities

    def maximization_step(self, X, responsibilities):
        N_k = responsibilities.sum(axis=0) + self.reg_covar  
        self.weights = N_k / (X.shape[0] + self.reg_covar)
        self.means = (responsibilities.T @ X) / N_k[:, np.newaxis]
        for k in range(self.n_components):
            diff = X - self.means[k]
            self.covariances[k] = (responsibilities[:, k, np.newaxis] * diff).T @ diff / N_k[k]
            self.covariances[k] += np.eye(X.shape[1]) * self.reg_covar 

    def log_likelihood(self, X):
        likelihoods = np.zeros((X.shape[0], self.n_components))
        for k in range(self.n_components):
            likelihoods[:, k] = self.weights[k] * self.gaussian_pdf(X, self.means[k], self.covariances[k])
        return np.sum(np.log(likelihoods.sum(axis=1) + self.reg_covar))  

    def fit(self, X):
        self.initialize_parameters(X)
        prev_log_likelihood = None
        
        for i in range(self.max_iters):
            responsibilities = self.expectation_step(X)
            self.maximization_step(X, responsibilities)
            log_likelihood = self.log_likelihood(X)
            
            if prev_log_likelihood is not None and abs(log_likelihood - prev_log_likelihood) < self.tol:
                break
            prev_log_likelihood = log_likelihood

    def predict(self, X):
        responsibilities = self.expectation_step(X)
        return np.argmax(responsibilities, axis=1)

X = df_scaled.drop(columns=['Cluster_sklearn']).values

gmm_custom = GaussianMixtureCustom(n_components=6)
gmm_custom.fit(X)

labels = gmm_custom.predict(X)

df_scaled['Cluster_GMM'] = labels

print(df_scaled.head())
```

## **Graficar clusters con elipses de covarianza**

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Ellipse

def plot_gmm_clusters(X, labels, means, covariances, feature_x, feature_y):
    plt.figure(figsize=(8, 6))
    
    scatter = plt.scatter(X[feature_x], X[feature_y], c=labels, cmap='viridis', alpha=0.5)
    
    for mean, cov in zip(means, covariances):
        eigvals, eigvecs = np.linalg.eigh(cov[:2, :2])  
        angle = np.degrees(np.arctan2(*eigvecs[:, 0][::-1]))
        
        width, height = 2 * np.sqrt(eigvals)
        ellipse = Ellipse(xy=mean[:2], width=width, height=height, angle=angle, edgecolor='red', fc='none', lw=2)
        plt.gca().add_patch(ellipse)

    plt.colorbar(scatter, label="Cluster")
    plt.xlabel(feature_x)
    plt.ylabel(feature_y)
    plt.title("Clusters de Mixture Models con Elipses de Covarianza")
    plt.show()

X = df_scaled.drop(columns=['Cluster_sklearn', 'Cluster_GMM']).values

feature_x = "CustAccountBalance"
feature_y = "TransactionAmount (INR)"

plot_gmm_clusters(df_scaled, df_scaled['Cluster_GMM'], gmm_custom.means, gmm_custom.covariances, feature_x, feature_y)
```

## **Implementación Mixture Models (con librerías)**

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score

features = ['CustAccountBalance', 'TransactionAmount (INR)', 'TransactionTime', 'CustGender', 'CustLocation']

gmm_sklearn = GaussianMixture(n_components=6, random_state=42, covariance_type='full')
gmm_sklearn.fit(df_scaled[features])

df_scaled['Cluster_GMM_sklearn'] = gmm_sklearn.predict(df_scaled[features])

log_likelihood_sklearn = gmm_sklearn.score(df_scaled[features]) * len(df_scaled)
print(f'Log-Likelihood del modelo GMM (sklearn): {log_likelihood_sklearn}')

df_sample = df_scaled.sample(frac=0.05, random_state=42)
silhouette_gmm_sklearn = silhouette_score(df_sample[features], df_sample['Cluster_GMM_sklearn'])

print(f'Coeficiente de Silueta para Mixture Models (sklearn, Muestra 5%): {silhouette_gmm_sklearn}')

def plot_gmm_clusters(X, labels, means, covariances, feature_x, feature_y):
    plt.figure(figsize=(8, 6))
    
    scatter = plt.scatter(X[feature_x], X[feature_y], c=labels, cmap='viridis', alpha=0.5)
    
    for mean, cov in zip(means, covariances):
        cov_2d = cov[:2, :2]  
        if cov_2d.shape == (2,2):  
            eigvals, eigvecs = np.linalg.eigh(cov_2d)  
            angle = np.degrees(np.arctan2(*eigvecs[:, 0][::-1]))
            
            width, height = 2 * np.sqrt(eigvals)
            ellipse = Ellipse(xy=mean[:2], width=width, height=height, angle=angle, edgecolor='red', fc='none', lw=2)
            plt.gca().add_patch(ellipse)

    plt.colorbar(scatter, label="Cluster")
    plt.xlabel(feature_x)
    plt.ylabel(feature_y)
    plt.title("Clusters de Mixture Models con sklearn y Elipses de Covarianza")
    plt.show()

feature_x = "CustAccountBalance"
feature_y = "TransactionAmount (INR)"

plot_gmm_clusters(df_scaled, df_scaled['Cluster_GMM_sklearn'], gmm_sklearn.means_, gmm_sklearn.covariances_, feature_x, feature_y)
```
