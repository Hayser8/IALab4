### 1. ¿Qué son los grafos computacionales y cuál es su importancia en el cálculo de gradientes en backpropagation?

Un **grafo computacional** es una representación estructurada de una función matemática donde los nodos representan operaciones o variables y las aristas representan el flujo de datos entre estas operaciones. En el contexto del cálculo de gradientes y el algoritmo de **backpropagation**, los grafos computacionales permiten descomponer la función de pérdida en una secuencia de operaciones elementales, facilitando el cálculo eficiente de derivadas mediante la **diferenciación automática**.

La importancia de los grafos computacionales en **backpropagation** radica en que permiten:

- **Optimización del cálculo de gradientes**: Evitan la redundancia en el cálculo de derivadas utilizando el método de la propagación inversa.
- **Eficiencia computacional**: Permiten almacenar los resultados intermedios de cada operación para ser reutilizados en la propagación inversa.
- **Facilidad en la implementación de modelos de aprendizaje profundo**: Frameworks como TensorFlow y PyTorch utilizan grafos computacionales para representar redes neuronales, facilitando la computación en hardware optimizado (GPU).

### 2. Componentes y pasos de una red neuronal. ¿Cómo mejorar el perceptrón del laboratorio pasado?

#### Componentes de una red neuronal:
Una red neuronal consta de los siguientes elementos fundamentales:

1. **Neuronas**: Unidades básicas que reciben una entrada, la procesan y generan una salida.
2. **Pesos y Sesgos**: Parámetros ajustables que permiten que la red aprenda patrones.
3. **Funciones de activación**: Aplicadas a la salida de las neuronas para introducir no linealidad (Ejemplo: ReLU, Sigmoide, Tanh).
4. **Capas**:
   - **Capa de entrada**: Recibe los datos.
   - **Capas ocultas**: Procesan la información mediante combinaciones lineales y funciones de activación.
   - **Capa de salida**: Genera la predicción final.
5. **Función de pérdida**: Mide la diferencia entre la salida real y la predicción.
6. **Optimizador**: Ajusta los pesos mediante **descenso de gradiente** para minimizar la pérdida.

#### Pasos en el entrenamiento de una red neuronal:
1. Inicializar los pesos y sesgos.
2. Realizar la **propagación hacia adelante** (Forward Pass) para obtener la salida.
3. Calcular la **función de pérdida**.
4. Aplicar **backpropagation** para calcular los gradientes.
5. **Actualizar los pesos** usando un optimizador como **Adam** o **SGD**.
6. Repetir los pasos hasta la convergencia.

#### Mejoras para el perceptrón del laboratorio pasado:
- **Agregar capas ocultas** para mejorar la capacidad de representación del modelo.
- **Usar una función de activación no lineal** (ej. ReLU) en lugar de una función escalón.
- **Implementar un optimizador avanzado** (como Adam) en lugar de solo usar descenso de gradiente estándar.
- **Normalización de datos** antes del entrenamiento para mejorar la convergencia.

### 3. Selección del valor K en K-Means usando el método de la silueta

El **método de la silueta** es una técnica utilizada para determinar el número óptimo de clusters (**K**) en el algoritmo de **K-Means** midiendo qué tan bien están agrupados los puntos dentro de sus clusters en comparación con otros clusters.

#### **Fórmula de la puntuación de silueta**:
Para cada punto \( i \) en el conjunto de datos:

1. **Calcular \( a(i) \)**: Distancia promedio entre \( i \) y todos los demás puntos en su mismo cluster.
2. **Calcular \( b(i) \)**: Distancia promedio entre \( i \) y los puntos en el cluster más cercano (al que no pertenece).
3. **Calcular la silueta de \( i \)**:
$$
s(i) = \frac{b(i) - a(i)}{\max(a(i), b(i))}
$$

   Donde:
   - \( s(i) \) varía entre -1 y 1.
   - Un valor cercano a 1 indica que el punto está bien agrupado.
   - Un valor cercano a 0 indica que el punto está en la frontera entre clusters.
   - Un valor negativo indica que el punto está mal agrupado.

#### **Elección de K**:
Se calcula la **puntuación media de silueta** para diferentes valores de \( K \), y se selecciona el \( K \) que maximiza esta puntuación.

**Asunciones**:
- Se asume que los datos pueden dividirse en clusters bien definidos.
- Se asume que la distancia euclidiana es una buena métrica para medir la cohesión de los clusters.

### 4. PCA y K-Means: ¿Cómo puede ayudar PCA a mejorar la calidad de los clusters?

**Principal Component Analysis (PCA)** es un método de reducción de dimensionalidad que transforma los datos a un nuevo espacio con menos dimensiones manteniendo la mayor variabilidad posible.

#### **Cómo ayuda PCA en K-Means**:
1. **Reducción de ruido y redundancia**: PCA elimina dimensiones poco informativas, lo que ayuda a K-Means a encontrar clusters más claros.
2. **Manejo de la "maldición de la dimensionalidad"**: K-Means puede tener problemas en espacios de alta dimensión. PCA reduce la dimensionalidad, facilitando la agrupación.
3. **Mejor visualización**: Permite proyectar los datos en 2D o 3D para visualizar mejor los clusters formados.
4. **Aceleración del algoritmo**: Con menos dimensiones, K-Means converge más rápido y consume menos recursos computacionales.

## **Preprocesamiento de datos**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# Cargar el dataset
file_path = "bank_transactions.csv"
df = pd.read_csv(file_path)

# Seleccionar columnas relevantes
df_cleaned = df[['CustAccountBalance', 'TransactionAmount (INR)', 'TransactionTime', 'CustGender', 'CustLocation']]

# Eliminar filas con valores nulos
df_cleaned = df_cleaned.dropna()

# Codificar variables categóricas
df_cleaned['CustGender'] = df_cleaned['CustGender'].map({'M': 0, 'F': 1})
df_cleaned['CustLocation'] = df_cleaned['CustLocation'].astype('category').cat.codes

# Escalar los datos
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_cleaned)
df_scaled = pd.DataFrame(df_scaled, columns=df_cleaned.columns)

# Mostrar primeras filas del dataset preprocesado
print(df_scaled.head())

# Verificar si hay NaN después de la estandarización
print(df_scaled.isna().sum())

# Si todavía hay valores NaN, eliminarlos
df_scaled = df_scaled.dropna()

# Verificar nuevamente
print(df_scaled.isna().sum())

```

## **Gráfica del codo**

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Definir el rango de clusters a probar
range_n_clusters = range(1, 11)

# Lista para guardar la inercia
inertia_values = []

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    inertia_values.append(kmeans.inertia_)

# Graficar el método del codo
plt.figure(figsize=(8, 5))
plt.plot(range_n_clusters, inertia_values, marker='o', linestyle='--')
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')
plt.title('Método del Codo para Determinar Número de Clusters')
plt.show()
```

### Análisis del Método del Codo

El método del codo se usa para determinar el número óptimo de clusters observando el punto donde la inercia deja de disminuir significativamente. En la gráfica obtenida, podemos notar que la curva tiene un descenso abrupto en los primeros valores de clusters y luego la reducción de la inercia se vuelve más gradual.

El punto óptimo en esta gráfica parece estar alrededor de **4 o 5 clusters**, ya que después de ese número la reducción de la inercia ya no es tan significativa. Si bien el número exacto puede variar dependiendo del análisis detallado, una elección común en estos casos es tomar el valor donde la curvatura cambia bruscamente.

Por lo tanto, procederemos con **K = 4 clusters** para el modelo de K-Means implementado sin librerías.

### Implementación de K-Means desde Cero

```python
import numpy as np

class KMeansCustom:
    def __init__(self, n_clusters, max_iters=100, tol=1e-4):
        self.n_clusters = n_clusters
        self.max_iters = max_iters
        self.tol = tol
        self.centroids = None

    def fit(self, X):
        np.random.seed(42)
        random_indices = np.random.choice(len(X), self.n_clusters, replace=False)
        self.centroids = X[random_indices]
        
        for _ in range(self.max_iters):
            clusters = self._assign_clusters(X)
            new_centroids = np.array([X[clusters == k].mean(axis=0) for k in range(self.n_clusters)])
            
            if np.linalg.norm(self.centroids - new_centroids) < self.tol:
                break
            
            self.centroids = new_centroids

    def _assign_clusters(self, X):
        distances = np.linalg.norm(X[:, np.newaxis] - self.centroids, axis=2)
        return np.argmin(distances, axis=1)
    
    def predict(self, X):
        return self._assign_clusters(X)

# Convertir el dataset a numpy array
X = df_scaled.values

# Entrenar modelo con K=4
kmeans_custom = KMeansCustom(n_clusters=5)
kmeans_custom.fit(X)

# Obtener etiquetas de los clusters
labels = kmeans_custom.predict(X)

# Agregar etiquetas al dataset
df_scaled['Cluster'] = labels

print(df_scaled.head())
```

### **Gráfico de dispersión de los clusters**

```python
import matplotlib.pyplot as plt

# Seleccionar dos características clave para visualizar los clusters
feature_x = "CustAccountBalance"
feature_y = "TransactionAmount (INR)"

# Graficar los clusters
plt.figure(figsize=(8, 6))
scatter = plt.scatter(df_scaled[feature_x], df_scaled[feature_y], c=df_scaled["Cluster"], cmap="viridis", alpha=0.5)
plt.colorbar(scatter, label="Cluster")
plt.xlabel(feature_x)
plt.ylabel(feature_y)
plt.title("Visualización de Clusters K-Means")
plt.show()
```

### **Evaluar la calidad de los clusters con el coeficiente de silueta**

```python
import pandas as pd
from sklearn.metrics import silhouette_score

# Tomar una muestra del 5% de los datos
df_sample = df_scaled.sample(frac=0.05, random_state=42)

# Calcular coeficiente de silueta en la muestra
silhouette_avg = silhouette_score(df_sample.drop(columns=['Cluster']), df_sample['Cluster'])
print(f'Coeficiente de Silueta (Muestra 5%): {silhouette_avg}')
```
